```python
predictions_path = '../predictions/para-dev-output.csv'
ground_truth_path = '../data/quora-dev.csv'
output_path = './para.png'
```

```python
# id	sentence1	sentence2	is_duplicate
p_id, p_duplicate = [], []
with open(predictions_path) as f:
    next(f)
    for line in f:
        id, is_dupli = line.strip().split(' , ')
        p_duplicate.append(float(is_dupli))
        p_id.append(id)


g_id, s1, s2, g_duplicate = [], [], [], []

with open(ground_truth_path) as f:
    next(f)
    for line in f:
        if len(line.strip().split('\t')) != 5:
            continue
        _, id, sen1, sen2, sim = line.strip().split('\t')
        g_duplicate.append(float(sim))
        g_id.append(id)
        s1.append(sen1)
        s2.append(sen2)
```

```python
num = 0
for i in range(len(p_id)):
    for j in range(len(g_id)):
        if p_id[i] == g_id[j] and p_duplicate[i] != g_duplicate[j]:
            print("{:.2f}".format(p_duplicate[i]) + " / " + "{:.2f}".format(g_duplicate[j]) + "\n" + s1[j] + "\n" + s2[j] + "\n")
            num += 1
            break
print(num)
            
```
